```python
import pandas as pd
```

```python
path = r'C:\Users\Cláudio Falcão\Google Drive\[01] - DATA SCIENCE\[05] - SCRIPTS\[01] - PYTHON\nba.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
# Retirando as duas primeiras colunas
colunas = ['Unnamed: 0','Unnamed: 0.1']
df = df.drop(colunas, axis=1)
df.head()
```

```python
# Tamanho do dataset
tamanho_df = len(df)
tamanho_df
```

```python
# Verificando os campos com dados nulos
soma_nulos = df.isnull().sum()
soma_nulos
```

```python
# Definir lista com campos nao nulos
coluna_nao_nulo = []
for col in df.columns:
    if df[col].isnull().sum() == 0:
        coluna_n_null = col
        coluna_nao_nulo.append(coluna_n_null)
print(coluna_nao_nulo)
```

```python
# Criando um novo dataset a partir dos campos não nulos
df_nao_nulo = df[coluna_nao_nulo]
df_nao_nulo.head()
```

```python
# Verificando 
df_nao_nulo.isnull().sum()
```

```python
# Exportando o novo dataset
caminho = r'C:\Users\Cláudio Falcão\Google Drive\[01] - DATA SCIENCE\[05] - SCRIPTS\[01] - PYTHON\nba_tratado.csv'
df_nao_nulo.to_csv(caminho)
```
