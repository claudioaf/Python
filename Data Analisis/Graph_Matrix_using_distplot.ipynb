```
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Loading dataset
housing = pd.read_csv('/content/drive/My Drive/Colab Notebooks/DATASET/housing.csv')
housing.head()
```

```
# Creating a list with the name of the columns
lista = housing.columns.tolist()
lista
```

```
# Deleting the ocean_proximity field name
del(lista[9])
lista
```

```
# Creating the probability density plots using a repeating structure

fig, ax = plt.subplots(3,3, figsize = (20,15))

# Defining the initial values for the parameters
count = 1
j = 0
k = 0

for i in lista:
  sns.distplot(housing[i],ax = ax[j, k])
  ax[j, k].set_xlim(min(housing[i]),max(housing[i]))
  ax[j, k].set_title('Distribution of '+i)
  ax[j, k].set_xlabel('')

  # defining the assignment of values to k and j 

  count = count + 1

  if count == 4:
    j = 1
  elif count == 7:
    j = 2

  k = k + 1

  if count == 4 or count == 7:
    k = 0

plt.show()
```
